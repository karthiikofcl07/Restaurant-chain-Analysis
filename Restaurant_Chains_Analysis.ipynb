<div class='section-header'>01 &nbsp;·&nbsp; ENVIRONMENT <span>&amp;</span> DATA LOADING</div>

```python
# ─────────────────────────────────────────────────────────────────────────────
# COGNIFYZ INTERNSHIP  |  Restaurant Chains Analysis
# Author : Karthikeyan Thirunavukkarasu
# ─────────────────────────────────────────────────────────────────────────────

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
from matplotlib.ticker import FuncFormatter
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
import plotly

# ── Global style ──────────────────────────────────────────────────────────────
pio.templates.default = 'plotly_white'

PALETTE = {
    'primary'   : '#c8102e',
    'secondary' : '#1a3a5c',
    'accent'    : '#b8860b',
    'light'     : '#f5f3ee',
    'dark'      : '#0a0a0f',
    'mid'       : '#6b7280',
}

PLOTLY_LAYOUT = dict(
    font=dict(family='IBM Plex Mono, monospace', size=12, color='#1a1a2e'),
    paper_bgcolor='#fafaf8',
    plot_bgcolor='#fafaf8',
    title_font=dict(family='IBM Plex Sans, sans-serif', size=18, color='#0a0a0f'),
    margin=dict(l=60, r=40, t=70, b=60),
    hoverlabel=dict(bgcolor='#0a0a0f', font_color='white', font_size=13),
)

print('✓ Libraries loaded successfully')
print(f'  pandas  {pd.__version__}  |  numpy {np.__version__}  |  plotly {plotly.__version__}')
```

```python
# ── Load & validate ───────────────────────────────────────────────────────────
df = pd.read_csv('data/cognifyz_dataset.csv')

# Normalise column names
df.columns = df.columns.str.strip()

print(f'Dataset shape  : {df.shape[0]:,} rows × {df.shape[1]} columns')
print(f'Null values    : {df.isnull().sum().sum()}')
print(f'Duplicate rows : {df.duplicated().sum()}')
print()
print('Schema overview')
print('─' * 52)
for col in df.columns:
    print(f'  {col:<26} {str(df[col].dtype):<10} | nunique={df[col].nunique()}')
```

<div class='section-header'>02 &nbsp;·&nbsp; CHAIN IDENTIFICATION</div>

```python
# ── Identify chains: any name appearing ≥ 2 times ─────────────────────────────
name_counts = df['Restaurant Name'].value_counts()
chain_names  = name_counts[name_counts >= 2].index
standalone   = name_counts[name_counts == 1].index

chain_df     = df[df['Restaurant Name'].isin(chain_names)].copy()
standalone_df= df[df['Restaurant Name'].isin(standalone)].copy()

print('═' * 52)
print(f'  CHAIN IDENTIFICATION SUMMARY')
print('═' * 52)
print(f'  Total restaurant records       : {len(df):>7,}')
print(f'  Unique restaurant names        : {df["Restaurant Name"].nunique():>7,}')
print(f'  ─────────────────────────────────────────────')
print(f'  Distinct chain brands (≥2)     : {len(chain_names):>7,}')
print(f'  Chain outlet records           : {len(chain_df):>7,}  ({len(chain_df)/len(df)*100:.1f}%)')
print(f'  ─────────────────────────────────────────────')
print(f'  Standalone restaurants         : {len(standalone_df):>7,}  ({len(standalone_df)/len(df)*100:.1f}%)')
print('═' * 52)
```

```python
# ── Chain statistics table ─────────────────────────────────────────────────────
chain_stats = (
    chain_df.groupby('Restaurant Name')
    .agg(
        Outlets        = ('Restaurant ID', 'count'),
        Avg_Rating     = ('Aggregate rating', 'mean'),
        Min_Rating     = ('Aggregate rating', 'min'),
        Max_Rating     = ('Aggregate rating', 'max'),
        Avg_Votes      = ('Votes', 'mean'),
        Total_Votes    = ('Votes', 'sum'),
        Cities_Present = ('City', 'nunique'),
        Avg_Cost       = ('Average Cost for two', 'mean'),
    )
    .reset_index()
    .sort_values('Outlets', ascending=False)
)
chain_stats['Avg_Rating'] = chain_stats['Avg_Rating'].round(2)
chain_stats['Avg_Votes']  = chain_stats['Avg_Votes'].round(0).astype(int)

print(f'Top 15 chains by outlet count:')
display(
    chain_stats.head(15)
    .style
    .set_table_styles([
        {'selector': 'thead th', 'props': 'background:#0a0a0f; color:white; font-weight:700; font-size:11px; letter-spacing:0.05em; padding:10px 14px;'},
        {'selector': 'tbody td', 'props': 'padding:9px 14px; font-size:12px; border-bottom:1px solid #e8e5df;'},
        {'selector': 'tbody tr:nth-child(odd)', 'props': 'background:#fafaf8;'},
    ])
    .bar(subset=['Outlets'],     color='#c8102e33')
    .bar(subset=['Avg_Rating'],  color='#1a3a5c33')
    .bar(subset=['Total_Votes'], color='#b8860b33')
    .format({'Avg_Rating': '{:.2f}', 'Avg_Cost': '{:,.0f}', 'Total_Votes': '{:,}'})
    .set_caption('Source: Cognifyz Restaurant Dataset | Chain = Restaurant Name appearing ≥ 2 times')
)
```

```python
# ── Fig 2: Chain vs Standalone proportion ─────────────────────────────────────
fig = go.Figure(go.Pie(
    labels=['Chain Outlets', 'Standalone Restaurants'],
    values=[len(chain_df), len(standalone_df)],
    hole=0.6,
    marker=dict(
        colors=[PALETTE['primary'], PALETTE['secondary']],
        line=dict(color='white', width=3)
    ),
    textinfo='label+percent',
    textfont=dict(size=13, family='IBM Plex Mono'),
    hovertemplate='%{label}<br>Count: %{value:,}<br>Share: %{percent}<extra></extra>',
))

fig.add_annotation(
    text='<b>9,551</b><br>Total',
    x=0.5, y=0.5, showarrow=False,
    font=dict(size=18, family='IBM Plex Sans', color='#0a0a0f'),
)

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='Market Composition: Chains vs Standalone', x=0.02),
    height=440,
    showlegend=True,
    legend=dict(orientation='h', y=-0.08, x=0.25, font=dict(size=12)),
)
fig.show()
```

```python
# ── Fig 3: Outlet size distribution ───────────────────────────────────────────
outlet_dist = chain_stats['Outlets'].value_counts().sort_index()

fig = go.Figure(go.Bar(
    x=outlet_dist.index,
    y=outlet_dist.values,
    marker=dict(
        color=outlet_dist.values,
        colorscale=[[0,'#d4e4f0'],[0.5,'#1a3a5c'],[1,'#c8102e']],
        showscale=False,
        line=dict(width=0)
    ),
    hovertemplate='Outlets: %{x}<br>Chains: %{y}<extra></extra>',
))

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='Distribution of Chain Sizes (Number of Outlets per Chain)', x=0.02),
    xaxis=dict(title='Number of Outlets', gridcolor='#e8e5df'),
    yaxis=dict(title='Number of Chains', gridcolor='#e8e5df'),
    height=400,
)
fig.show()

print(f'\n  Chains with 2 outlets        : {(chain_stats["Outlets"]==2).sum()}')
print(f'  Chains with 3-9 outlets      : {((chain_stats["Outlets"]>=3)&(chain_stats["Outlets"]<=9)).sum()}')
print(f'  Chains with 10-49 outlets    : {((chain_stats["Outlets"]>=10)&(chain_stats["Outlets"]<=49)).sum()}')
print(f'  Chains with 50+ outlets      : {(chain_stats["Outlets"]>=50).sum()}')
```

<div class='section-header'>04 &nbsp;·&nbsp; RATINGS ANALYSIS <span>&amp;</span> QUALITY BENCHMARKING</div>

```python
# ── Top 15 highest-rated chains (≥5 outlets for statistical validity) ──────────
qualified = chain_stats[chain_stats['Outlets'] >= 5].sort_values('Avg_Rating', ascending=False)
top_rated  = qualified.head(15).sort_values('Avg_Rating')

fig = go.Figure()

fig.add_trace(go.Bar(
    x=top_rated['Avg_Rating'],
    y=top_rated['Restaurant Name'],
    orientation='h',
    name='Avg Rating',
    marker=dict(
        color=top_rated['Avg_Rating'],
        colorscale=[[0,'#f0e6d3'],[0.4,'#b8860b'],[1,'#c8102e']],
        showscale=True,
        colorbar=dict(title='Rating', thickness=12, len=0.7),
        line=dict(width=0)
    ),
    text=top_rated['Avg_Rating'].round(2).astype(str),
    textposition='outside',
    textfont=dict(size=11, family='IBM Plex Mono', color='#0a0a0f'),
    hovertemplate='<b>%{y}</b><br>Avg Rating: %{x:.2f}<extra></extra>',
))

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='Top 15 Highest-Rated Chains (Min. 5 Outlets — Statistically Valid)', x=0.02),
    xaxis=dict(title='Average Aggregate Rating', range=[3.5, 5.0], gridcolor='#e8e5df'),
    yaxis=dict(title=''),
    height=520,
    showlegend=False,
)

fig.add_vline(x=4.0, line=dict(dash='dot', color=PALETTE['secondary'], width=1.5),
              annotation_text='4.0 threshold', annotation_position='top right',
              annotation_font=dict(size=10, color=PALETTE['secondary']))
fig.show()
```

```python
# ── Chain vs Standalone rating distribution comparison ────────────────────────
# Exclude 0-rated (no review) for distribution clarity
chain_rated    = chain_df[chain_df['Aggregate rating'] > 0]['Aggregate rating']
standalone_rated = standalone_df[standalone_df['Aggregate rating'] > 0]['Aggregate rating']

fig = go.Figure()

fig.add_trace(go.Histogram(
    x=chain_rated, name='Chain Outlets',
    nbinsx=30, opacity=0.8,
    marker=dict(color=PALETTE['primary'], line=dict(width=0.5, color='white')),
    histnorm='percent',
))
fig.add_trace(go.Histogram(
    x=standalone_rated, name='Standalone',
    nbinsx=30, opacity=0.65,
    marker=dict(color=PALETTE['secondary'], line=dict(width=0.5, color='white')),
    histnorm='percent',
))

fig.add_vline(x=chain_rated.mean(), line=dict(dash='dash', color=PALETTE['primary'], width=2),
              annotation_text=f'Chain μ={chain_rated.mean():.2f}',
              annotation_position='top left',
              annotation_font=dict(size=10, color=PALETTE['primary']))
fig.add_vline(x=standalone_rated.mean(), line=dict(dash='dash', color=PALETTE['secondary'], width=2),
              annotation_text=f'Standalone μ={standalone_rated.mean():.2f}',
              annotation_position='top right',
              annotation_font=dict(size=10, color=PALETTE['secondary']))

fig.update_layout(
    **PLOTLY_LAYOUT,
    barmode='overlay',
    title=dict(text='Rating Distribution: Chain Outlets vs Standalone (Rated Only)', x=0.02),
    xaxis=dict(title='Aggregate Rating', gridcolor='#e8e5df'),
    yaxis=dict(title='% of Restaurants', gridcolor='#e8e5df'),
    height=420,
    legend=dict(orientation='h', y=1.1, x=0.5, xanchor='center'),
)
fig.show()

print(f'  Chain outlets    — Mean rating: {chain_rated.mean():.3f}  | Median: {chain_rated.median():.2f}  | Std: {chain_rated.std():.3f}')
print(f'  Standalone       — Mean rating: {standalone_rated.mean():.3f}  | Median: {standalone_rated.median():.2f}  | Std: {standalone_rated.std():.3f}')
```

```python
# ── Rating tier breakdown per chain ───────────────────────────────────────────
def rating_tier(r):
    if r == 0:    return 'Not Rated'
    elif r < 2.5: return 'Poor (<2.5)'
    elif r < 3.5: return 'Average (2.5–3.5)'
    elif r < 4.0: return 'Good (3.5–4.0)'
    elif r < 4.5: return 'Very Good (4.0–4.5)'
    else:         return 'Excellent (≥4.5)'

chain_df['Rating_Tier'] = chain_df['Aggregate rating'].apply(rating_tier)

tier_order  = ['Excellent (≥4.5)','Very Good (4.0–4.5)','Good (3.5–4.0)','Average (2.5–3.5)','Poor (<2.5)','Not Rated']
tier_colors = [PALETTE['primary'],'#2e7d32','#388e3c','#f9a825',PALETTE['mid'],'#bdbdbd']
tier_map    = dict(zip(tier_order, tier_colors))

tier_counts = (
    chain_df['Rating_Tier']
    .value_counts()
    .reindex(tier_order, fill_value=0)
)

fig = go.Figure(go.Pie(
    labels=tier_counts.index.tolist(),
    values=tier_counts.values.tolist(),
    hole=0.45,
    marker=dict(colors=tier_colors, line=dict(color='white', width=2)),
    textinfo='label+percent',
    textfont=dict(size=11, family='IBM Plex Mono'),
    direction='clockwise',
    sort=False,
))

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='Rating Tier Distribution Across Chain Outlets', x=0.02),
    height=460,
    legend=dict(orientation='v', y=0.5, font=dict(size=11)),
)
fig.show()
```

```python
# ── City-level chain concentration ────────────────────────────────────────────
city_chain = (
    chain_df.groupby('City')
    .agg(
        chain_outlets     = ('Restaurant ID', 'count'),
        distinct_chains   = ('Restaurant Name', 'nunique'),
        avg_chain_rating  = ('Aggregate rating', 'mean'),
    )
    .sort_values('chain_outlets', ascending=False)
    .head(12)
    .reset_index()
)

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=['Chain Outlets per City', 'Distinct Chain Brands per City'],
    horizontal_spacing=0.1,
)

fig.add_trace(go.Bar(
    x=city_chain['City'], y=city_chain['chain_outlets'],
    marker=dict(color=PALETTE['primary'], line=dict(width=0)),
    text=city_chain['chain_outlets'],
    textposition='outside', textfont=dict(size=10),
    hovertemplate='%{x}<br>Outlets: %{y}<extra></extra>',
), row=1, col=1)

fig.add_trace(go.Bar(
    x=city_chain['City'], y=city_chain['distinct_chains'],
    marker=dict(color=PALETTE['secondary'], line=dict(width=0)),
    text=city_chain['distinct_chains'],
    textposition='outside', textfont=dict(size=10),
    hovertemplate='%{x}<br>Brands: %{y}<extra></extra>',
), row=1, col=2)

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='Geographic Concentration of Restaurant Chains — Top 12 Cities', x=0.02),
    height=440,
    showlegend=False,
)
fig.update_xaxes(tickangle=30)
fig.show()
```

```python
# ── City reach of top chains (heatmap-style) ───────────────────────────────────
top10_chains = chain_stats.head(10)['Restaurant Name'].tolist()
top10_df     = chain_df[chain_df['Restaurant Name'].isin(top10_chains)]

pivot = (
    top10_df.groupby(['Restaurant Name', 'City'])
    .size()
    .reset_index(name='Outlets')
    .pivot(index='Restaurant Name', columns='City', values='Outlets')
    .fillna(0)
)

# Keep top cities by total outlets
top_cities = top10_df['City'].value_counts().head(12).index.tolist()
pivot = pivot.reindex(columns=top_cities, fill_value=0)

fig = go.Figure(go.Heatmap(
    z=pivot.values,
    x=pivot.columns.tolist(),
    y=pivot.index.tolist(),
    colorscale=[[0,'#f5f3ee'],[0.3,'#d4e4f0'],[0.7,'#1a3a5c'],[1,'#c8102e']],
    text=pivot.values.astype(int),
    texttemplate='%{text}',
    textfont=dict(size=10, family='IBM Plex Mono'),
    hovertemplate='<b>%{y}</b> in <b>%{x}</b><br>Outlets: %{z}<extra></extra>',
    showscale=True,
    colorbar=dict(title='Outlets', thickness=12),
))

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='City × Chain Outlet Heatmap — Top 10 Chains across Top Cities', x=0.02),
    xaxis=dict(title='City', tickangle=30),
    yaxis=dict(title='Chain', autorange='reversed'),
    height=440,
)
fig.show()
```

```python
# ── Comparative performance: Premium chains vs Fast-food chains ────────────────
premium  = ['Barbeque Nation', 'Farzi Cafe', "Chili's", 'Punjab Grill', 'Mocha']
fastfood = ["Domino's Pizza", 'Subway', "McDonald's", 'Pizza Hut', 'KFC']

prem_df  = chain_df[chain_df['Restaurant Name'].isin(premium)]
fast_df  = chain_df[chain_df['Restaurant Name'].isin(fastfood)]

metrics = {
    'Category': ['Premium Casual', 'Fast Food'],
    'Outlets'  : [len(prem_df), len(fast_df)],
    'Avg Rating': [prem_df['Aggregate rating'].mean().round(3), fast_df['Aggregate rating'].mean().round(3)],
    'Avg Votes' : [prem_df['Votes'].mean().round(0), fast_df['Votes'].mean().round(0)],
    'Cities'    : [prem_df['City'].nunique(), fast_df['City'].nunique()],
}

comp_df = pd.DataFrame(metrics)

cats = ['Outlets', 'Avg Rating', 'Avg Votes', 'Cities']
prem_vals = [comp_df.loc[0, c] for c in cats]
fast_vals = [comp_df.loc[1, c] for c in cats]

# Normalise for radar-style comparison
norm_prem = [v / max(v, f) for v, f in zip(prem_vals, fast_vals)]
norm_fast = [v / max(v, f) for v, f in zip(fast_vals, prem_vals)]

print('Premium Casual vs Fast Food Chains — Performance Benchmark')
print('─' * 60)
for c in cats:
    pv = comp_df.loc[0, c]
    fv = comp_df.loc[1, c]
    winner = 'Premium' if pv > fv else 'Fast Food'
    print(f'  {c:<14}: Premium={pv:<10.1f} FastFood={fv:<10.1f} → {winner}')

print()
display(
    comp_df.set_index('Category')
    .style
    .set_table_styles([
        {'selector': 'thead th', 'props': 'background:#1a3a5c; color:white; font-size:12px; padding:10px 16px;'},
        {'selector': 'tbody td', 'props': 'padding:9px 16px; font-size:13px; font-weight:600;'},
        {'selector': 'tbody tr:first-child', 'props': 'background:#fff5f5;'},
    ])
    .format({'Avg Rating': '{:.3f}', 'Avg Votes': '{:.0f}'})
    .highlight_max(axis=0, color='#e8f5e9')
)
```

```python
# ── Correlation: outlets vs avg rating ────────────────────────────────────────
corr_df = chain_stats[chain_stats['Outlets'] >= 2].copy()
corr_val = corr_df[['Outlets','Avg_Rating','Total_Votes','Cities_Present']].corr()

fig = go.Figure(go.Heatmap(
    z=corr_val.values.round(3),
    x=corr_val.columns.tolist(),
    y=corr_val.index.tolist(),
    colorscale=[[0,'#c8102e'],[0.5,'#fafaf8'],[1,'#1a3a5c']],
    zmid=0,
    text=corr_val.values.round(3),
    texttemplate='%{text}',
    textfont=dict(size=13, family='IBM Plex Mono'),
    hovertemplate='%{y} × %{x}<br>r = %{z}<extra></extra>',
    showscale=True,
    colorbar=dict(title='r', thickness=12),
))

fig.update_layout(
    **PLOTLY_LAYOUT,
    title=dict(text='Correlation Matrix: Scale, Rating, Votes, Geographic Reach', x=0.02),
    height=380,
)
fig.show()

r_outlets_rating = corr_df[['Outlets','Avg_Rating']].corr().iloc[0,1]
print(f'\n  Pearson r (Outlets × Avg_Rating) = {r_outlets_rating:.4f}')
print(f'  Interpretation: {"Positive" if r_outlets_rating > 0 else "Negative"} correlation — scale {"associates with" if abs(r_outlets_rating) < 0.3 else "strongly predicts"} rating (weak)' if abs(r_outlets_rating) < 0.3 else '')
```
